```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('tips')

# ----------------------------------------)---------
# 1. HISTOGRAMS
# -------------------------------------------------
plt.hist(df['total_bill'])
plt.title("Distribution of Total Bill")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
plt.show()

plt.hist(df['tip'])
plt.title("Distribution of Tips")
plt.xlabel("Tip Amount")
plt.ylabel("Frequency")
plt.show()

# -------------------------------------------------
# 2. BAR CHARTS
# -------------------------------------------------
```

```python

# Average tip by day
avg_tip_day = df.groupby('day')['tip'].mean()
avg_tip_day.plot(kind='bar')
plt.title("Average Tip by Day")
plt.xlabel("Day")
plt.ylabel("Average Tip")
plt.show()

# Average total_bill Lunch vs Dinner
avg_bill_time = df.groupby('time')['total_bill'].mean()
avg_bill_time.plot(kind='bar')
plt.title("Average Total Bill: Lunch vs Dinner")
plt.xlabel("Time")
plt.ylabel("Average Total Bill")
plt.show()

#
```

```python

# 3. SCATTER PLOT
# -------------------------------------------------
# Smokers vs Non-smokers highlighted using color
colors = df['smoker'].map({'Yes': 'red', 'No': 'blue'})

plt.scatter(df['total_bill'], df['tip'], c=colors)
plt.title("Total Bill vs Tip (Smoker Highlight)")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.legend(handles=[
    plt.Line2D([], [], marker='o', color='red', linestyle='', label='Smoker'),
    plt.Line2D([], [], marker='o', color='blue', linestyle='', label='Non-Smoker')
])
plt.show()

```

```python
sns.boxplot(x='day', y='total_bill', data=df)
plt.title('Boxplot of Total Bill by Day')
plt.show()
```

```python
sns.boxplot(data=df, x='sex', y='tip')
plt.show()
```
