```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.datasets import make_classification
```

```python
df = pd.read_csv('DA314_S8_EmployeeAttrition_Data_Practice.csv')
```

<h1>boxplot to detect Monthlylncome and
TotalWorkingYears</h1>

```python
boxplot_names = ['MonthlyIncome' , 'TotalWorkingYears']
fig, axes = plt.subplots(1 , 2 , figsize=(16 , 8))
sns.boxplot(data=df , x = boxplot_names[0] , ax=axes[0] , color='lightcoral')
sns.boxplot(data=df , x = boxplot_names[1] , ax=axes[1] , color='grey')
plt.suptitle('Boxplot-MonthlyIncome & TotalWorkingYears')
plt.show()
```

```python
sns.boxplot(data = df , x = 'YearsAtCompany' , color = '#ADD8E6')
plt.title('YearsAtCompany')
plt.show()
```

```python
p95 = df['MonthlyIncome'].quantile(0.95)
sns.boxplot(x = df[df['MonthlyIncome'] <= p95]['MonthlyIncome'] , color = 'lightcoral' , width = 0.8)
plt.xlabel('MonthlyIncome_Capped')
```
