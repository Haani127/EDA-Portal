```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load your GlamNest data
df = pd.read_csv('P1.csv')

# Set style for better visualizations
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (15, 6)

# ========== FINAL AMOUNT VISUALIZATIONS ==========
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# Histogram with KDE for final_amount
sns.histplot(df['final_amount'], bins=30, color='skyblue', edgecolor='black',
             alpha=0.7, kde=True, ax=axes[0])
axes[0].set_xlabel('Final Amount', fontsize=12)
axes[0].set_ylabel('Frequency', fontsize=12)
axes[0].set_title('Distribution of Final Amount (Histogram)', fontsize=14, fontweight='bold')
axes[0].grid(axis='x', alpha=0.5)

# Boxplot for final_amount
axes[1].boxplot(df['final_amount'], vert=True, patch_artist=True,
                boxprops=dict(facecolor='lightblue', color='blue'),
                whiskerprops=dict(color='blue'),
                capprops=dict(color='blue'),
                medianprops=dict(color='red', linewidth=2))
axes[1].set_ylabel('Final Amount', fontsize=12)
axes[1].set_title('Distribution of Final Amount (Boxplot)', fontsize=14, fontweight='bold')
axes[1].grid(axis='y', alpha=0.5)

plt.tight_layout()
plt.savefig('final_amount_analysis.png', dpi=300, bbox_inches='tight')
plt.show()

# ========== SERVICE RATING VISUALIZATIONS ==========
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# Histogram with KDE for service_rating
sns.histplot(df['service_rating'], bins=20, color='lightcoral', edgecolor='black',
             alpha=0.7, kde=True, ax=axes[0])
axes[0].set_xlabel('Service Rating', fontsize=12)
axes[0].set_ylabel('Frequency', fontsize=12)
axes[0].set_title('Distribution of Service Rating (Histogram)', fontsize=14, fontweight='bold')
axes[0].grid(axis='x', alpha=0.5)

# Boxplot for service_rating
axes[1].boxplot(df['service_rating'], patch_artist=True,
                boxprops=dict(facecolor='lightcoral', color='red'),
                whiskerprops=dict(color='red'),
                capprops=dict(color='red'),
                medianprops=dict(color='darkred', linewidth=2))
axes[1].set_ylabel('Service Rating', fontsize=12)
axes[1].set_title('Distribution of Service Rating (Boxplot)', fontsize=14, fontweight='bold')
axes[1].grid(axis='y', alpha=0.5)

plt.tight_layout()
plt.savefig('service_rating_analysis.png', dpi=300, bbox_inches='tight')
plt.show()
```
