```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('P3.csv')
```

# Heatmap

```python
plt.figure(figsize=(8, 7))
sns.heatmap(df[['price', 'discount%', 'discounted_amount(INR)', 'final_amount', 'service_rating']].corr(), annot=True, cmap='coolwarm')
plt.title("Coorelation Heatmap")
plt.show()
```
