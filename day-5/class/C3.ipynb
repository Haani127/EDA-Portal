```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('C3.csv')
df
```

```python
colms = df[['avg_cost_for_two', 'rating']]
```

```python
sns.heatmap(colms.corr(), annot=True, cmap='coolwarm')
plt.show()
```

```python
sns.regplot(data=colms, x='avg_cost_for_two', y='rating', color='orange', line_kws={'color':'red', 'linestyle':'--'})
plt.show()
```
