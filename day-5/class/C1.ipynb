```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('C1.csv')
df
```

# Histogram

```python
sns.histplot(df['total_amount'], bins=[500, 1000, 1500, 2000, 2500, 3000, 3500], color='skyblue', edgecolor='black', kde=True)
plt.grid(axis='y', linestyle='--', color='red')
plt.show()
```

# BOX plot

```python
sns.boxplot(df, x='total_amount', color='brown')
plt.grid(axis='x', linestyle='--', color='skyblue')
plt.show()
```
