```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('P3.csv')
df.head()
df.columns
```

```python

plt.figure(figsize=(6, 5))
sns.countplot(
    data=df,
    x='OverTime',
    hue='Attrition',
    palette='Blues'
)
plt.title('Attrition by Overtime', fontsize=14)
plt.xlabel('OverTime')
plt.ylabel('Count')
plt.legend(title='Attrition')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(7, 5))
sns.countplot(
    data=df,
    x='Department',
    hue='Attrition',
    palette='Blues'
)
plt.title('Attrition by Department', fontsize=14)
plt.xlabel('Department')
plt.ylabel('Count')
plt.legend(title='Attrition')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 5))
sns.countplot(
    data=df,
    x='JobSatisfaction',
    hue='Attrition',
    palette='Blues'
)
plt.title('Attrition by Job Satisfaction', fontsize=14)
plt.xlabel('Job Satisfaction (1=Low, 4=High)')
plt.ylabel('Count')
plt.legend(title='Attrition')
plt.tight_layout()
plt.show()
```

```python
bins = [18, 25, 35, 45, 55, 65]
labels = ['18-25', '26-35', '36-45', '46-55', '56-65']
df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

plt.figure(figsize=(7, 5))
sns.countplot(
    data=df,
    x='AgeGroup',
    hue='Attrition',
    palette='Blues'
)
plt.title('Attrition by Age Group', fontsize=14)
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.legend(title='Attrition')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(7, 5))
sns.boxplot(
    data=df,
    x='Attrition',
    y='MonthlyIncome',
    hue='Attrition',
    palette='Blues'
)
plt.title('Monthly Income vs Attrition', fontsize=14)
plt.xlabel('Attrition')
plt.ylabel('Monthly Income')
plt.tight_layout()
plt.show()
```
