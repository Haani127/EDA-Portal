```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_tips = sns.load_dataset('tips')
df_wine = pd.read_csv('wine.csv')
```

```python
print(df_wine.head())
print(df_wine.describe())

# Select only numeric features
numeric_cols = df_wine.select_dtypes(include=['float64', 'int64']).columns

# Histogram for each feature
for col in numeric_cols:
    plt.figure()
    plt.hist(df_wine[col])
    plt.title(f"Histogram of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()

# Boxplot for each feature
for col in numeric_cols:
    plt.figure()
    plt.boxplot(df_wine[col])
    plt.title(f"Boxplot of {col}")
    plt.ylabel(col)
    plt.show()

# Density Plot (Distribution) for each feature
for col in numeric_cols:
    plt.figure()
    df_wine[col].plot(kind='density')
    plt.title(f"Density Plot of {col}")
    plt.xlabel(col)
    plt.show()
```

```python

# Load Tips datase

print(df_tips.head())
print(df_tips.describe())

# Scatter plot: total_bill vs tip
plt.figure()
plt.scatter(df_tips['total_bill'], df_tips['tip'])
plt.title("Total Bill vs Tip")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()

# Boxplot: sex vs tip
plt.figure()
df_tips.boxplot(column="tip", by="sex")
plt.title("Tip Amount by Gender")
plt.suptitle("")   # Removes extra title
plt.xlabel("Gender")
plt.ylabel("Tip")
plt.show()

# Bar plot: day vs average total_bill
avg_bill_by_day = df_tips.groupby('day')['total_bill'].mean()

plt.figure()
avg_bill_by_day.plot(kind='bar')
plt.title("Average Total Bill by Day")
plt.xlabel("Day")
plt.ylabel("Average Total Bill")
plt.show()

# Correlation between total_bill and tip
correlation = df_tips['total_bill'].corr(df_tips['tip'])
print("Correlation between Total Bill and Tip:", correlation)
```
