```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('P3.csv')
df.head()
```

```python
shift_emp = df.groupby('Shift')['EmployeeID'].count().reset_index()

plt.bar(shift_emp['Shift'], shift_emp['EmployeeID'], color='orange')
plt.xticks(range(0, 4))
plt.show()
```

```python
df.columns
```

```python
role = df.groupby('JobRole')['EmployeeID'].count().reset_index().sort_values('EmployeeID', ascending=False)
plt.bar(role['JobRole'], role['EmployeeID'])
plt.show()
```

```python
marr = df.groupby('MaritalStatus')['EmployeeID'].count().reset_index()
plt.pie(marr['EmployeeID'], labels=marr['MaritalStatus'], autopct='%1.1f%%')
plt.show()
```

```python
wbf = df.groupby('WorkLifeBalance')['EmployeeID'].count().reset_index()

plt.bar(wbf['WorkLifeBalance'], wbf['EmployeeID'])
plt.show()
```

```python
sns.histplot(
    data=df,
    x='PercentSalaryHike',
    kde=True,
    cbar='blue',
    # edge_color='black'
)
plt.show()
```
